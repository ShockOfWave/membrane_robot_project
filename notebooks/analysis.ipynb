```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
os.chdir(os.path.expanduser('..'))
plt.style.use('ggplot')
```

```python
df = pd.ExcelFile('data/raw/Book1.xlsx')
df.sheet_names
```

```python
df_nacl = pd.read_excel(df, df.sheet_names[0])
df_licl = pd.read_excel(df, df.sheet_names[1])
df_cscl = pd.read_excel(df, df.sheet_names[2])
df_kcl = pd.read_excel(df, df.sheet_names[3])
df_cscl_kcl = pd.read_excel(df, df.sheet_names[4])
```

### NaCl

```python
df_nacl = df_nacl.loc[2:]
df_nacl
```

```python
plt.plot(df_nacl[df_nacl.columns[0]].values, df_nacl[df_nacl.columns[1]].values, label='pH feed')
plt.plot(df_nacl[df_nacl.columns[0]].values, df_nacl[df_nacl.columns[3]].values, label='pH premeate 1')
plt.plot(df_nacl[df_nacl.columns[0]].values, df_nacl[df_nacl.columns[5]].values, label='pH premeate 2')
plt.legend()
plt.title('Time/pH for 0.1M NaCl dataset')
plt.xlabel('Time (s)')
plt.ylabel('pH')
plt.savefig('reports/figures/time_ph_nacl.svg', format='svg', dpi=1200)
```

```python
plt.plot(df_nacl[df_nacl.columns[1]].values, df_nacl[df_nacl.columns[2]].values, label='Conductivity/pH feed')
plt.plot(df_nacl[df_nacl.columns[3]].values, df_nacl[df_nacl.columns[4]].values, label='Conductivity/pH premeate 1')
plt.plot(df_nacl[df_nacl.columns[5]].values, df_nacl[df_nacl.columns[6]].values, label='Conductivity/pH premeate 2')
plt.legend()
plt.title('pH/Conductivity for 0.1M NaCl dataset')
plt.xlabel('pH')
plt.ylabel('Conductivity')
plt.savefig('reports/figures/ph_conductivity_nacl.svg', format='svg', dpi=1200)
```

```python
plt.plot(df_nacl[df_nacl.columns[0]].values, df_nacl[df_nacl.columns[2]].values, label='Conductivity feed')
plt.plot(df_nacl[df_nacl.columns[0]].values, df_nacl[df_nacl.columns[4]].values, label='Conductivity premeate 1')
plt.plot(df_nacl[df_nacl.columns[0]].values, df_nacl[df_nacl.columns[6]].values, label='Conductivity premeate 2')
plt.legend()
plt.title('Time/Conductivity for 0.1M NaCl dataset')
plt.xlabel('Time (s)')
plt.ylabel('Conductivity')
plt.savefig('reports/figures/time_conductivity_nacl.svg', format='svg', dpi=1200)
```

```python
fig, axs = plt.subplots(5, 2, figsize=(12, 20))

fig.suptitle('Data subplot', fontsize=24)

axs[0, 0].plot(df_kcl[df_kcl.columns[0]].values, df_kcl[df_kcl.columns[2]].values)
axs[0, 0].plot(df_kcl[df_kcl.columns[0]].values, df_kcl[df_kcl.columns[4]].values)
axs[0, 0].set_title('0.1M KCl', fontsize=14)
axs[0, 0].set(ylabel='Conductivity', xlabel='Time')

axs[0, 1].plot(df_kcl[df_kcl.columns[0]].values, df_kcl[df_kcl.columns[1]].values)
axs[0, 1].plot(df_kcl[df_kcl.columns[0]].values, df_kcl[df_kcl.columns[3]].values)
axs[0, 1].set_title('0.1M KCl', fontsize=14)
axs[0, 1].set(ylabel='pH', xlabel='Time')

axs[1, 0].plot(df_nacl[df_nacl.columns[0]].values, df_nacl[df_nacl.columns[2]].values)
axs[1, 0].plot(df_nacl[df_nacl.columns[0]].values, df_nacl[df_nacl.columns[4]].values)
axs[1, 0].set_title('0.1M NaCl', fontsize=14)
axs[1, 0].set(ylabel='Conductivity', xlabel='Time')

axs[1, 1].plot(df_nacl[df_nacl.columns[0]].values, df_nacl[df_nacl.columns[1]].values)
axs[1, 1].plot(df_nacl[df_nacl.columns[0]].values, df_nacl[df_nacl.columns[3]].values)
axs[1, 1].set_title('0.1M NaCl', fontsize=14)
axs[1, 1].set(ylabel='pH', xlabel='Time')

axs[2, 0].plot(df_licl[df_licl.columns[0]].values, df_licl[df_licl.columns[2]].values)
axs[2, 0].plot(df_licl[df_licl.columns[0]].values, df_licl[df_licl.columns[4]].values)
axs[2, 0].set_title('0.1M LiCl', fontsize=14)
axs[2, 0].set(ylabel='Conductivity', xlabel='Time')

axs[2, 1].plot(df_licl[df_licl.columns[0]].values, df_licl[df_licl.columns[1]].values)
axs[2, 1].plot(df_licl[df_licl.columns[0]].values, df_licl[df_licl.columns[3]].values)
axs[2, 1].set_title('0.1M LiCl', fontsize=14)
axs[2, 1].set(ylabel='pH', xlabel='Time')

axs[3, 0].plot(df_cscl[df_cscl.columns[0]].values, df_cscl[df_cscl.columns[2]].values)
axs[3, 0].plot(df_cscl[df_cscl.columns[0]].values, df_cscl[df_cscl.columns[4]].values)
axs[3, 0].set_title('0.1M CsCl', fontsize=14)
axs[3, 0].set(ylabel='Conductivity', xlabel='Time')

axs[3, 1].plot(df_cscl[df_cscl.columns[0]].values, df_cscl[df_cscl.columns[1]].values)
axs[3, 1].plot(df_cscl[df_cscl.columns[0]].values, df_cscl[df_cscl.columns[3]].values)
axs[3, 1].set_title('0.1M CsCl', fontsize=14)
axs[3, 1].set(ylabel='pH', xlabel='Time')

axs[4, 0].plot(df_cscl_kcl[df_cscl_kcl.columns[0]].values, df_cscl_kcl[df_cscl_kcl.columns[2]].values)
axs[4, 0].plot(df_cscl_kcl[df_cscl_kcl.columns[0]].values, df_cscl_kcl[df_cscl_kcl.columns[4]].values)
axs[4, 0].set_title('0.1M CsCl + 0.1M KCl', fontsize=14)
axs[4, 0].set(ylabel='Conductivity', xlabel='Time')

axs[4, 1].plot(df_cscl_kcl[df_cscl_kcl.columns[0]].values, df_cscl_kcl[df_cscl_kcl.columns[1]].values)
axs[4, 1].plot(df_cscl_kcl[df_cscl_kcl.columns[0]].values, df_cscl_kcl[df_cscl_kcl.columns[3]].values)
axs[4, 1].set_title('0.1M CsCl + 0.1M KCl', fontsize=14)
axs[4, 1].set(ylabel='pH', xlabel='Time')
```
