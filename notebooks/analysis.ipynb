```python
df = pd.ExcelFile('data/raw/Book1.xlsx')
df.sheet_names
```

```python
plt.rcParams.update({'font.size': 18})
fig, axs = plt.subplots(5, 2, figsize=(14, 32))

fig.suptitle('Data subplot', fontsize=24)

axs[0, 0].plot(df_kcl[df_kcl.columns[0]].values, df_kcl[df_kcl.columns[2]].values, label='Feed')
axs[0, 0].plot(df_kcl[df_kcl.columns[0]].values, df_kcl[df_kcl.columns[4]].values, label='Premeate')
axs[0, 0].set_title('0.1M KCl', fontsize=18)
axs[0, 0].legend(fontsize='18')
axs[0, 0].set(ylabel='Conductivity (mS/cm)', xlabel='Time (s)')

axs[0, 1].plot(df_kcl[df_kcl.columns[0]].values, df_kcl[df_kcl.columns[1]].values, label='Feed')
axs[0, 1].plot(df_kcl[df_kcl.columns[0]].values, df_kcl[df_kcl.columns[3]].values, label='Premeate')
axs[0, 1].set_title('0.1M KCl', fontsize=18)
axs[0, 1].legend(fontsize='18')
axs[0, 1].set(ylabel='pH', xlabel='Time (s)')

axs[1, 0].plot(df_nacl[df_nacl.columns[0]].values, df_nacl[df_nacl.columns[2]].values, label='Feed')
axs[1, 0].plot(df_nacl[df_nacl.columns[0]].values, df_nacl[df_nacl.columns[4]].values, label='Premeate')
axs[1, 0].set_title('0.1M NaCl', fontsize=18)
axs[1, 0].legend(fontsize='18')
axs[1, 0].set(ylabel='Conductivity (mS/cm)', xlabel='Time (s)')

axs[1, 1].plot(df_nacl[df_nacl.columns[0]].values, df_nacl[df_nacl.columns[1]].values, label='Feed')
axs[1, 1].plot(df_nacl[df_nacl.columns[0]].values, df_nacl[df_nacl.columns[3]].values, label='Premeate')
axs[1, 1].set_title('0.1M NaCl', fontsize=18)
axs[1, 1].legend(fontsize='18')
axs[1, 1].set(ylabel='pH', xlabel='Time (s)')

axs[2, 0].plot(df_licl[df_licl.columns[0]].values, df_licl[df_licl.columns[2]].values, label='Feed')
axs[2, 0].plot(df_licl[df_licl.columns[0]].values, df_licl[df_licl.columns[4]].values, label='Premeate')
axs[2, 0].set_title('0.1M LiCl', fontsize=18)
axs[2, 0].legend(fontsize='18')
axs[2, 0].set(ylabel='Conductivity (mS/cm)', xlabel='Time (s)')

axs[2, 1].plot(df_licl[df_licl.columns[0]].values, df_licl[df_licl.columns[1]].values, label='Feed')
axs[2, 1].plot(df_licl[df_licl.columns[0]].values, df_licl[df_licl.columns[3]].values, label='Premeate')
axs[2, 1].set_title('0.1M LiCl', fontsize=18)
axs[2, 1].legend(fontsize='18')
axs[2, 1].set(ylabel='pH', xlabel='Time (s)')

axs[3, 0].plot(df_cscl[df_cscl.columns[0]].values, df_cscl[df_cscl.columns[2]].values, label='Feed')
axs[3, 0].plot(df_cscl[df_cscl.columns[0]].values, df_cscl[df_cscl.columns[4]].values, label='Premeate')
axs[3, 0].set_title('0.1M CsCl', fontsize=18)
axs[3, 0].legend(fontsize='18')
axs[3, 0].set(ylabel='Conductivity (mS/cm)', xlabel='Time (s)')

axs[3, 1].plot(df_cscl[df_cscl.columns[0]].values, df_cscl[df_cscl.columns[1]].values, label='Feed')
axs[3, 1].plot(df_cscl[df_cscl.columns[0]].values, df_cscl[df_cscl.columns[3]].values, label='Premeate')
axs[3, 1].set_title('0.1M CsCl', fontsize=18)
axs[3, 1].legend(fontsize='18')
axs[3, 1].set(ylabel='pH', xlabel='Time (s)')

axs[4, 0].plot(df_cscl_kcl[df_cscl_kcl.columns[0]].values, df_cscl_kcl[df_cscl_kcl.columns[2]].values, label='Feed')
axs[4, 0].plot(df_cscl_kcl[df_cscl_kcl.columns[0]].values, df_cscl_kcl[df_cscl_kcl.columns[4]].values, label='Premeate')
axs[4, 0].set_title('0.1M CsCl + 0.1M KCl', fontsize=18)
axs[4, 0].legend(fontsize='18')
axs[4, 0].set(ylabel='Conductivity (mS/cm)', xlabel='Time (s)')

axs[4, 1].plot(df_cscl_kcl[df_cscl_kcl.columns[0]].values, df_cscl_kcl[df_cscl_kcl.columns[1]].values, label='Feed')
axs[4, 1].plot(df_cscl_kcl[df_cscl_kcl.columns[0]].values, df_cscl_kcl[df_cscl_kcl.columns[3]].values, label='Premeate')
axs[4, 1].set_title('0.1M CsCl + 0.1M KCl', fontsize=18)
axs[4, 1].legend(fontsize='18')
axs[4, 1].set(ylabel='pH', xlabel='Time (s)')

fig.tight_layout()
fig.savefig('reports/figures/data_subplot.svg', format='svg', dpi=1200)
fig.savefig('reports/figures/data_subplot.png', format='png', dpi=1200)
```
