```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
xls = pd.ExcelFile('../data/raw/Book1.xlsx')
print(xls.sheet_names)

df_nacl = pd.read_excel(xls, xls.sheet_names[0])
df_licl = pd.read_excel(xls, xls.sheet_names[1])
df_cscl = pd.read_excel(xls, xls.sheet_names[2])
df_kcl = pd.read_excel(xls, xls.sheet_names[3])
df_cscl_kcl = pd.read_excel(xls, xls.sheet_names[4])
```

### 0.1M NaCl dataframe

```python
df_nacl
```

##### Cut first 2 rows

```python
df_nacl = df_nacl.loc[2:]
df_nacl = df_nacl.dropna(axis=0)
```

##### Get only inputs and outputs for model

```python
X_data_1 = df_nacl['Conductivity.1'].values
X_data_2 = df_nacl['Conductivity.2'].values
y_data = df_nacl['pH'].values
X_train_1, X_test_1, X_train_2, X_test_2, y_train, y_test = train_test_split(X_data_1, X_data_2, y_data, test_size=0.2)
```

##### Create dataframe

```python
df_nacl_final_train = pd.DataFrame(data={
    'Conductivity_premeate_1': X_train_1,
    'Conductivity_premeate_2': X_train_2,
    'pH_feed': y_train
})

df_nacl_final_test = pd.DataFrame(data={
    'Conductivity_premeate_1': X_test_1,
    'Conductivity_premeate_2': X_test_2,
    'pH_feed': y_test
})
```

```python
df_nacl_final_train
```

##### Save dataframe

```python
df_nacl_final_train.to_csv('../data/processed/data_nacl_train.csv', index=False)
df_nacl_final_test.to_csv('../data/processed/data_nacl_test.csv', index=False)
```
